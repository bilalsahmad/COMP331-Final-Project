```python
!pip install ucimlrepo

import pandas as pd
from ucimlrepo import fetch_ucirepo

# Fetch dataset
adult = fetch_ucirepo(id=2)
X = adult.data.features

# How many ? values per column
question_mark_counts = (X == "?").sum()

# Percentagae of ? per column
question_mark_percentage = (question_mark_counts / 32509 * 100)

# Combine into a single results table
results = pd.DataFrame({
    "Question Mark Count": question_mark_counts,
    "Percentage (%)": question_mark_percentage.round(2)
})

print(results)
```
