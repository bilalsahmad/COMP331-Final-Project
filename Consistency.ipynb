```python
!pip install ucimlrepo
import pandas as pd
from ucimlrepo import fetch_ucirepo

# Fetch dataset
adult = fetch_ucirepo(id=2)
X = adult.data.features

# Correct mapping for the Adult dataset
edu_map = {
    "Preschool": 1,
    "1st-4th": 2,
    "5th-6th": 3,
    "7th-8th": 4,
    "9th": 5,
    "10th": 6,
    "11th": 7,
    "12th": 8,
    "HS-grad": 9,
    "Some-college": 10,
    "Assoc-voc": 11,
    "Assoc-acdm": 12,
    "Bachelors": 13,
    "Masters": 14,
    "Prof-school": 15,
    "Doctorate": 16
}

# Compare education and education-num
matches = X["education"].map(edu_map) == X["education-num"]

# Count mismatches
num_mismatches = (~matches).sum()

# Percentage
percentage_mismatch = (num_mismatches / len(X)) * 100

print("Mismatches:", num_mismatches)
print("Percentage mismatched: {:.2f}%".format(percentage_mismatch))
```
